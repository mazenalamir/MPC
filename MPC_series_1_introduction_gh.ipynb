## This course 

This course is an introduction to an advanced control design framework called 

**Model Predictive Control**. 

This name comes from the fact that this methodology is based on **Predicting** the future behaviour of the system using its **Dynamic Model** in order to choose the best **control** to be applied at each instant. 

This first notebook help understanding this simple idea.
